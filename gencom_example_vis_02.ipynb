```python
# import external data

quakes = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/earthquakes-23k.csv')
```

We can check the content of the dataframe, and see what the titles of the columns are:

```python
quakes.columns
```

We can create some statistical plots using the seaborn library.

To start, we can create some basic overview plots.

```python
sns.pairplot(quakes).savefig("fig01.png")
```

```python
sns.relplot(data=quakes,
           x='Longitude',
           y='Latitude',
           hue="Magnitude",
            size="Magnitude",
           alpha=0.6,
            palette="viridis_r")
```

```python
sns.jointplot(data=quakes,
              x='Longitude',
            y='Latitude',
              color="m",)
```

We can then start to analyse the data in a bit more detail

```python
sns.set_theme(style="ticks")
sns.jointplot(data=quakes,
              x='Longitude',
            y='Latitude',
              color="m",
              kind="hex")
```

```python
print(quakes.Date)
```

```python
# Convert the 'Date' column to datetime format
quakes['Date'] = pd.to_datetime(quakes['Date'], errors='coerce')  # invalid formats become NaT

# Extract the month
quakes['Month'] = quakes['Date'].dt.month
```

```python
# this one will take a while!
sns.relplot(
    data=quakes,
    x='Longitude',y='Latitude',
    hue="Magnitude",
    col="Month", col_wrap=4,
)
```

```python
sns.set_theme(style="ticks")
sns.jointplot(data=quakes,
              x='Latitude',
            y='Month',
              color="m",
              kind="hex")
```
